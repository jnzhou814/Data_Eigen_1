```python
#导入包
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
```

```python
#文件路径
filepath="D://Market_Basket_Optimisation.csv"
```

```python
#读取文件*
raw_data = pd.read_csv(filepath, header=None)
```

```python
#转换为dataframe
words_df = DataFrame(raw_data)
```

```python
#将数据中的所有词生成list并返回
words_list = []
for i in range(0, words_df.shape[0]):
    for j in range(0, raw_data.shape[1]):
        word_item = words_df.values[i, j]
        if str(word_item) !='nan':
            words_list.append(str(word_item))
```

```python
# 统计词频
freq_dict = nltk.FreqDist(words_list)
```

```python
#按照降序排列
new_dict = sorted(freq_dict.items(), key=lambda x: x[1], reverse=True)
```

```python
#查看前10位
new_dict[:10]
```

```python
wc = WordCloud(
        max_words=100,
        width=2000,
        height=1200,
    )
```

```python
# 将商品名称组成text，以空格分隔
cut_text = " ".join(words_list)
    # 生成词云并保存
word_cloud = wc.generate(cut_text)
```

```python
plt.imshow(word_cloud)
plt.axis("off")
```
